```python
import pandas as pd

df = pd.read_csv("hadoop-YARN-5085.csv")
df
```

```python
df = df.dropna(axis=1)
df
```

```python
import os
df = df.assign(basename=df['Name'].apply(lambda x: os.path.basename(x)))
df
```

```python
df = df.drop(columns = ['Kind', 'Name'], axis = 1)
df
```

```python
last_column_index = len(df.columns) - 1
df = pd.concat([df.iloc[:, last_column_index:], df.iloc[:, :-1]], axis=1)

# rename the first column to "Name"
df = df.rename(columns={df.columns[0]: "Name"})
df
```

```python
df.to_csv('product_metrics.csv', index=False)
```
